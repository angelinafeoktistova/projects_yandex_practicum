### Апартаменты, студия и сводобная планировка

В столбцах is_apartment, studio, open_plan содержатся данные типа object и bool о том, является ли объект недвижимости апартаментами/студией/квартирой со свободной планировкой. Познакомимся с данными в этих столбцах. 

Так как информацию заполняли пользователи, предположу, что пропуски были допущены там, где на вопрос пользователь подразумевал ответ "нет".

```python
# подготовка и вызов функции для подсчета пропусков и процента пропусков в столбце 
# я буду вызывать эту функцию повторно при анализе каждого столбца
# вывод количества уникальных значений в столбцах

def show_percent_nan(column):
    number_nan = data[column].isna().sum()
    percent_nan = round((number_nan / len(data)), 2)
    print(f'Пропущено {number_nan} значений, это {percent_nan:.2%}.')


display(data['is_apartment'].value_counts().reset_index())
show_percent_nan('is_apartment')

display(data['studio'].value_counts().reset_index())
show_percent_nan('studio')

display(data['open_plan'].value_counts().reset_index())
show_percent_nan('open_plan')
```

```python
# замена пропусков в столбце по апартаментам на False
# замена булевых значений на строки
# проверка, что не осталось пропущенных значений в апартаментах

data['is_apartment'] = data['is_apartment'].fillna(False).replace({True: 'да', False: 'нет'})
data['studio'] = data['studio'].replace({True: 'да', False: 'нет'})
data['open_plan'] = data['open_plan'].replace({True: 'да', False: 'нет'})
display(data['is_apartment'].value_counts().reset_index())
show_percent_nan('is_apartment')
```

**Вывод**

**Пропуски**  Пропуски могли возникнуть по причине того, что продавцы не знали, что именно нужно указывать в ячейках. Также пропуски могли возникнуть в связи с тем, что продавцы подразумевали отрицательный ответ. 
Пропущенные значения в столбце "is_apartment", составили 88%, они были заполнены. В связи с тем, что пропусков в данном столбце очень много, его не следует учитывать при дальнейшем анализе. 

<br/>
Рекомендация для усовершенствования системы: предлагать продавцам возможные однозначные варианты ответа: да/нет (подсказки); ограничить возможность оставлять ячейки не заполненными.

<br/>

**Аномалии**  Аномалии в столбцах не обнаружены. Подавляющее большинство объектов в архиве - квартиры. Апартаментов в таблице очень мало, как и студий, и квартир со свободной планировкой.



### Балконы

В столбце balcony содержатся данные типа float64 о количестве балконов в квартире. Посмотрим на значения столбца, обработаем пропущенные значения и выявим аномалии.

```python
# просмотр уникальных значений столбца
# вызов функции с подсчетом пропусков и процентов пропусков в столбце

display(sorted(data['balcony'].unique()))
show_percent_nan('balcony')
```

В столбце много пропусков, есть квартиры с подозрительно большим количеством балконов.
<br/> Вероятно, что значения были пропущены пользователями из-за того, что балконов в квартирах нет. Заменим все пропущенные значения на 0.

<br/> Также нужно изменить тип данных. Балконов может быть только целое число. Не подходящий тип данных усложнит чтение графиков.

### Потолок

В столбце ceiling_height содержатся данные типа float64 о высоте потолка в квартире. Посмотрим на  значения столбца, обработаем пропуски и выявим аномалии.

```python
# вызов функции с подсчетом пропусков и процентов пропусков в столбце
# вывод числового описания столбца
# вывод среза  с аномальной высотой потолка меньше 2 метров
# вывод уникальных значений столбца с высотой больше 3 метров

show_percent_nan('ceiling_height')
display(data['ceiling_height'].describe())
display(data.query('ceiling_height < 2').copy())
sorted(data[data['ceiling_height'] > 3]['ceiling_height'].unique())
```

Как видим, в столбце много пропусков. Кроме того, у квартир есть аномальные значения высоты потолка. Есть потолки ниже 2 метров и потолки высотой 32 метра. Полагаю, что такие значения указаны с ошибкой. Указанные аномалии будут обработаны позднее.
Сейчас высчитаем медиану и заполним пропуски.

```python
# получаем все записи датафрейма, где значения у потолка являются нормальными (от 2 до 3 метров)
# вычисляем медиану нормального потолка
normal_height = data.loc[(data['ceiling_height'] >= 2) & (data['ceiling_height'] <= 3)]
normal_height_median = round(normal_height['ceiling_height'].median(), 1)


# с помощью цикла заполняем словарь с медианами нормального потолка для каждого дома определенной этажности  
floors_medians = {}
for floor in data['floors_total'].unique():
    range_height = normal_height.loc[normal_height['floors_total'] == floor].copy()
    if len(range_height) > 0:
        floors_medians[floor] = round(range_height['ceiling_height'].median(), 1)
    else:
        floors_medians[floor] = normal_height_median
              
# временно заменяем все пропуски на 0
data['ceiling_height'] = data['ceiling_height'].fillna(0)


# создаем и вызываем функцию по заполнению пропусков значениями из словаря медиан
def fill_nan_ceiling(row):
    height = row['ceiling_height']
    floors_total = row['floors_total']
    if height == 0:
        return floors_medians[floors_total]
    else:
        return height

data['ceiling_height'] = data.apply(fill_nan_ceiling, axis=1)


# проверяем, что все пропуски заполнены
show_percent_nan('ceiling_height')
```

### Населенный пункт

В столбце locality_name содержатся данные типа object с именем населенного пункта.

```python
# вызов функции с подсчетом пропусков и процентов пропусков в столбце
# вывод числового описания столбца
# просмотр всех уникальных значений столбца

show_percent_nan('locality_name')
data.dropna(subset=['locality_name'], inplace=True)
sorted(data['locality_name'].unique())
```

```python
# обработка столбца будет произведена в несколько этапов

# замена буквы "ё" в наименовании
# унификация наименования всех населенных пунктов (останутся только: деревня, поселок, СНТ)

data['locality_name'] = (
    data['locality_name']
    .str.replace('ё', 'е')
    .replace(['городской поселок ', 'коттеджный поселок ', 'поселок городского типа ', \
                  'поселок при железнодорожной станции ', 'поселок станции ', 'село '], 'поселок ', regex=True)
    .replace(['садоводческое некоммерческое товарищество ', 'садовое товарищество '], 'СТ ', regex=True)    
)


# создание нового столбца 'locality_only_name', который будет содержать только названия населенного пункта

data['locality_only_name'] = (
    data['locality_name']
    .replace(['поселок ', 'деревня ', 'СТ ', ], '', regex=True)
)


# создание нового столбца 'locality_type', который будет содержать только категорию населенного пункта 
# там, где категория не была указана изначально, будет стоять "не указано"

def get_locality_type(full_name):
    if full_name.startswith ('поселок') or full_name.startswith('деревня') or full_name.startswith('СТ'):
        split_locality = full_name.split()
        locality_type = split_locality[0]
        return locality_type
    else:
        return 'не указано'


data['locality_type'] = data['locality_name'].apply(get_locality_type)

# проверка получившихся столбцов

display(data['locality_type'].unique())
display(data['locality_only_name'].value_counts().reset_index().head(10))
display(data[['locality_type', 'locality_name', 'locality_only_name']].head(10))
```

Отдельно стоит обратить внимание, что в изначальном столбце были указаны населенные пункты с одинаковым наименованием, но разным типом. Например, деревня Федоровское и поселок Федоровское. ## Обработка аномалий

На основе выявленных аномалий, а также гистограмм и диаграмм размаха очистим данные от выбросов и аномалий для дальнейшего анализа и выявления зависимостей. 

```python
# обработка аномалий балконов
# удаление строк с тремя и более балконами

good_data = data.loc[data['balcony'] < 3].copy()

# обработка аномалий потолков
# создаем и вызываем функцию, чтобы передвинуть точку, т.к. это ошибка при внесении данных (вместо 25 будет 2,5)
# отбрасываем строки, в которых высота потолка осталась аномальной (срез)

def move_ceiling_dot(row):
    height = row['ceiling_height']
    if height >= 20:
        value = round(height/10, 1)
        return value
    else:
        return round(height,1)

good_data['ceiling_height'] = good_data.apply(move_ceiling_dot, axis=1)


good_data = good_data.query('ceiling_height >= 2 and ceiling_height <= 3')
    
    
# обработка аномалий общей площади
# отбрасываем строки, в которых общая площадь сильно выбивается на основе гистограммы (срез)

good_data = good_data.query('total_area >= 12 and total_area <= 150')



# обработка аномалий жилой площади
# удаляем очевидно аномальные значения - строки с квартирами, в которых: 
# 1) жилая площадь меньше или равна 8 м, но при этом есть несколько комнат 
# 2) жилая площадь меньше 6 метров
# 3) жилая площадь с кухней больше общей площади квартиры (должно быть как минимум столько же)

good_data = good_data.drop(good_data.query('living_area < 8 and rooms > 1 or living_area < 6').index)
good_data = good_data.drop(good_data.query('kitchen_area + living_area > total_area').index)
                 
                 
                 
# обработка аномалий кухни
# удаляем очевидные ошибочные значения, где площадь кухни меньше 3 метров

good_data = good_data.drop((good_data.query('kitchen_area < 3').index)) 
                 

    
# обработка аномалий аэропорта
# удаление слишком близко расположенных к аэропорту квартир
good_data = good_data.drop(good_data.query('airports_nearest < 1').index)


# обработка аномалий стоимости
# удаление слишком выбивающихся значений
good_data = good_data.drop(good_data.query('last_price < 100000 or last_price > 25000000').copy().index)


# обработка аномалий комнат
# удаление квартиры с 19 комнатами
# удаление квартир с 0 комнат и не являющихся студией/апартаментами/квартирой со свободной планировкой
# удаление квартир с большим количеством комнат и маленькой жилой площадью

good_data = good_data.drop(good_data.query('rooms == rooms.max()').index)
good_data = good_data.drop(good_data.query('rooms == 0 and studio =="нет" and is_apartment =="нет" and open_plan == "нет"').index)
good_data = good_data.drop(good_data.query('rooms > 6 and living_area < 50').index)
display(good_data.head(10))

print(f'Общее количество строк после обработки пропусков и аномалий составляет {len(good_data)}.')
```

Посмотрим на числовое описание данных столбцов и на гистограммы столбцов, получившихся после обработки пропусков и удаления явных и выбивающихся аномалий.

Для исследования зависимости между параметрами и стоимостью объекта я буду использовать данные, очищенные от разного рода аномалий и выбросов.

#### Зависимость между общей площадью и полной стоимостью объекта

```python
# подготовка функции для созадния сводной таблицы по изучаемому фактору
# в функции устанавливаю критерий по количеству объявлений более 10, т.к. меньшее количество
# считаю недопустимым учитывать при выявлении зависимостей - слишком маленькая выборка


def get_good_pivot_table(column_for_analysis):
    table = good_data.pivot_table(index=column_for_analysis, values='last_price', 
                                  aggfunc=['count', 'median']).reset_index()
    table.columns = [column_for_analysis, 'count_ads', 'median']
    table = table.query('count_ads > 10').sort_values('median', ascending = False)
    return table

# подготовка функции для вывода подсчитанного коэф. Пирсона по изучаемому фактору
# я буду переиспользовать обе функции
def print_corr (column_for_analysis, table):
    print( 'Коэффицент корреляции Пирсона составляет: ', 
          round(table[column_for_analysis].corr(table['median']),2))
```

```python
# вызов функции построения сводной таблицы с общей площадью
# вызов функции расчета коэф. Пирсона
# построение диаграммы


corr_total_area = get_good_pivot_table('total_area')
display(corr_total_area.head(10))
corr_total_area.plot(y='median', style='o', grid=True, 
                     alpha=0.7, title='Зависимость общей площади и стоимости квартир', 
                     figsize=(8, 5), legend=None)

plt.xlabel('Общая площадь')
plt.ylabel('Стоимость квартир')

print_corr('total_area', corr_total_area)
```

**Вывод**

Существует сильная линейная связь между общей площадью квартиры и ее стоимостью. Коэффициент Пирсона положительный и составляет 0,95. 

Между величинами положительная корреляция и прямая зависимость: с увеличением площади растет стоимость квартиры.

На основании получившихся данных были сделаны общие выводы о состоянии рынка продажи квартир за исследуемый период.

Для исследования аномалии и выбивающиеся значения были обработаны. Явные аномалии удалены, подготовлен отдельный датафрейм с данными, "очищенными" от аномалий. Именно этот датафрейм использовался на стадии поиска зависимостей между данными.

После обработки всех аномалий, выбивающихся значений и пропусков, там где это было возможно, был проведен анализ данных. В частности, были проделаны следующие действия:

- анализ сроков продажи квартир,
- определение зависимости между разными параметрами квартир и ее стоимостью,
- вычисление средней цены одного квадратного метра в самых популярных населенных пунктах, 
- определение зависимости стоимости квартиры от удаленности от центра города в Санкт-Петербурге.

Для удобства просмотра полученных результатов в исследовании использовались графики.
