# Приоритизация гипотез и проведение A/B-тестирования

В рамках данного исследования произведена приоритизация гипотез и проанализированы результаты A/B-тестирования для выявления возможностей по увеличению выручки интернет-магазина. 

**Цель** Приоритизация гипотез для выявления тех из них, которые следует проверить в первую очередь. Анализ результатов A/B-теста для принятия решения по нему: остановить тест и зафиксировать итог либо продолжить тест.

В частности, необходимо:

*Приоритизация гипотез*

- применить фреймворк ICE для приоритизации гипотез,
- применить фреймворк RICE для приоритизации гипотез,
- проанализировать и сравнить результаты по ICE и RICE. 

*Анализ результатов A/B-теста*

- проанализировать кумулятивную выручку по группам и построить соответствующий график,
- проанализировать средний чек по группам, а также его относительное изменение и построить соответствующие графики,
- проанализировать среднее число заказов на посетителя по группам, а также ее относительное изменение и построить соответствующие графики,
- проанализировать выбросы и всплески в данных: построить точечные графики, посчитать перцентили, выбрать границы для определения аномальных заказов по из числу на посетителя и стоимости заказов,
- посчитать статистическую значимость различий в среднем числе заказов на посетителя по группам по сырым и очищенным данным,
- посчитать статистическую значимость различий в среднем чеке заказа по группам по сырым и очищенным данным,
- принять решение по результатам теста: остановить тест и зафиксировать итог либо продолжить тест.


**Описание данных** Для исследования предоставлены три файла формата `csv`:

- файл с 9 гипотезами по увеличению выручки с расставленными оценками по 10-ти бальной шкале по критериям Reach, Impact, Confidence, Effort,
- файл с данными о заказах: ID заказа и ID пользователя, дата заказа, выручка заказа и группа A/B-теста, в которую попал заказ,
- файл с данными о пользователях: дата, группа A/B-теста и количество пользователей в указанную дату в указанной группе теста.

**Содержание исследования** 

Исследование проведено в несколько этапов.

На первом этапе проведена работа по приоритизации гипотез. Проведена предобработка файла с гипотезами. ОСуществлена приоритизация по ICE и сделаны выводы. Затем проведена приоритизация по RICE и по полученным результатам сделаны выводы. Результаты двух приоритизаций были сопоставлены и подведены итоги сравнения применения фреймворков.

На втором этапе произведен анализ результатов A/B-теста. Проведена предобработка данных, выявлены аномальные пользователи, а также произведена подготовка кумулятивных данных для дальнейшего анализа. Далее произведен анализ кумулятивных метрик по группам тестирования: выручка, средний чек, среднее количество заказов на посетителя - построены соответствующие графики и выдвинуты предположения. Затем проведен анализ выбросов и всплесков: построены точечные графики числа заказов, стоимости заказов, а также вычислены перцентили - выявлены пользователи с аномальным поведениям. После были рассчитаны статистические значимости различий метрик по сырым и очищенным данным, сделаны выводы. Также отдельно рассмотрена статистическая значимость различий среднего чека по данным без пользователей, попавших одновременно в обе тестируемые группы. 
На основе полученных результатов было принято решение о заверешнии тестирования и сделан вывод, что эксперимент положительно влияет на среднее количество заказов на посетителя и не влияет на показатели среднего чека.

Есть и гипотезы с макисмально низкой оценкой уверенности, например, *"Изменить цвет фона главной страницы, чтобы увеличить вовлеченность пользователей"* и *"Добавить страницу отзывов клиентов о магазине, что позволит увеличить количество заказов"*.

`Efforts (затраты):` Есть гипотеза, которая требует большого количества затрат - *"Запустить собственную службу доставки, что сократит срок доставки заказов"*, что вполне очевидно из ее формулировки.

При таком общем анализе гипотез, до их приоритазации с помощью фреймворков можно предположить:

- гипотеза *"Запустить собственную службу доставки, что сократит срок доставки заказов"* требует очень высоких затрат, но достаточно малый охват. Возможно, такую гипотезу стоит проверять в конце очереди.
- гипотеза *"Изменить цвет фона главной страницы, чтобы увеличить вовлеченность пользователей"* также имеет достаточно низкие оценки по всем параметрам. Возможно, такую гипотезу стоит проверять в конце очереди.

**Проверим оценочные предположения с помощью фреймвороков.**

### Приоритизация гипотез с помощью ICE и RICE

Сначала приоритизируем гипотезы с помощью  фреймворка ICE, который учитывает такие характеристики как: Impact, Confidence, Efforts. Но этот фреймворк игнорирует Reach.  Затем приоритизируем гипотезы с применением фреймворка RICE, который уже учитывает и параметр Reach.  После сравним результаты.


ICE рассчитывается по формуле: Impact  * Confidence / Efforts

RICE рассчитывается по формуле: Reach * Impact  * Confidence / Efforts

```python
# подготовим функцию для применения фреймворков и ранжирования получившегося приоритета гипотезы 
def get_framework_score(df, name_col_score, name_col_priority):
    df_score = df.copy()
    if name_col_score == 'Ice_score':
        df_score[name_col_score] = round(((df_score['Impact'] * df_score['Confidence']) / df_score['Efforts']),2)
    else:
        df_score[name_col_score] = round(((df_score['Reach'] * df_score['Impact'] 
                                           * df_score['Confidence']) / df_score['Efforts']),2)
    df_score = df_score.sort_values(by=name_col_score, ascending=False).reset_index(drop=True)
    df_score[name_col_priority] = range(1, len(df_score) + 1)
    return df_score
```

## Анализ кумулятивных данных

Для дальнейшего анализа результатов теста необходимо аккумулировать и подготовить данные, чтобы отследить изменение результатов эксперимента на каждый день тестирования и наглядно посмотреть различия между группами.

```python
# дневная агрерация заказов
orders_daily = (orders.groupby(['date', 'group'], as_index=False)
                    .agg(transactions=('transactionId', 'nunique'),
                         revenue=('revenue', 'sum'))
                    .sort_values(['group', 'date'])
                )

#дневная агрегация клиентов
visitors_daily= (visitors.groupby(['date', 'group'], as_index=False)
                        .agg(visitors=('visitors', 'sum'))
                        .sort_values(['group', 'date'])
                )


orders_daily['transactions_cum'] = orders_daily.groupby('group')['transactions'].cumsum()
orders_daily['revenue_cum'] = orders_daily.groupby('group')['revenue'].cumsum()

visitors_daily['visitors_cum'] = visitors_daily.groupby('group')['visitors'].cumsum()

cummulative_data = orders_daily.merge(visitors_daily, on=['date','group']).sort_values(['date', 'group'])

# добавим столбец с конверсией
cummulative_data['avg_number_transactions'] = cummulative_data['transactions_cum'] / cummulative_data['visitors_cum']



# разделим кумалитивные данные по тестируемым группам
cummulative_a = cummulative_data.query('group == "A"')
cummulative_b = cummulative_data.query('group == "B"')

# создадим объединенный датасет с разделенными по столбцам данными по группам
merged_cummulative_data = cummulative_a.merge(cummulative_b, on=['date'], suffixes=['_a', '_b'])
merged_cummulative_data.head()
```

**Вывод**
Мы подготовили данные для дальнейшего удобного построения графиков.
В частности, мы получили таблицы с кумулятивными данными по дате с учетом группы тестирования: количество заказов, коилчество покупателей, сумма выручки и количество посетителей.

## Статистическая значимость различий в метриках

### Статистическая значимость различий в среднем количестве заказов на посетителя между группами по «сырым» данным

Посчитаем статистическую значимость различий в среднем числе заказов на посетителя между группами без удаления аномальных данных. 


**Нулевая гипотеза**: Различий между средним числом заказов нет нет.

**Альтернативная гипотеза**: Среднее число заказов в группах различается.

```python
# подготовим данные для подсчетов

# рассчитаем отдельно для каждой группы число пользователей на каждый день
visitors_daily_a = visitors.query('group == "A"') [['date', 'visitors']]
visitors_daily_b = visitors.query('group == "B"') [['date', 'visitors']]
visitors_daily_a.columns = ['date', 'visitors_per_day_a']
visitors_daily_b.columns = ['date', 'visitors_per_day_b']

# подсчитаем отдельно для каждой группы количество заказов и общую стоимость на каждый день
orders_daily_a = (orders.query('group == "A"')
                        .groupby('date', as_index=False)
                        .agg(transactions=('transactionId', 'nunique'),
                             revenue=('revenue', 'sum'))
                 )

orders_daily_b = (orders.query('group == "B"')
                        .groupby('date', as_index=False)
                        .agg(transactions=('transactionId', 'nunique'),
                             revenue=('revenue', 'sum'))
                 )

orders_daily_a.columns = ['date', 'orders_per_date_a', 'revenue_per_date_a']
orders_daily_b.columns = ['date', 'orders_per_date_b', 'revenue_per_date_b']


# объединим данные
data_raw = (orders_daily_a.merge(orders_daily_b, on='date')
                          .merge(visitors_daily_a, on='date')
                          .merge(visitors_daily_b, on='date'))
data_raw.head()
```

```python
# рассчитаем p_value и относительное различие в среднем числе заказов на посетителя по сырым данным

total_visitors_a = data_raw['visitors_per_day_a'].sum()

# число пользователей, которые что-то купили
buyers_a = len(orders_by_users_a)

# заполним нулями для пользователей, которые ничего не купили 
zeros_a = pd.Series([0] * (total_visitors_a - buyers_a), name='orders')


total_visitors_b = data_raw['visitors_per_day_b'].sum()
buyers_b = len(orders_by_users_b)
zeros_b = pd.Series([0] * (total_visitors_b - buyers_b), name='orders')

sample_a = pd.concat([orders_by_users_a['orders'], zeros_a], ignore_index=True)
sample_b = pd.concat([orders_by_users_b['orders'], zeros_b], ignore_index=True)

p_value = stats.mannwhitneyu(sample_a, sample_b).pvalue

print('\np-value по "сырым данным":', "{0:.5f}".format(p_value))

if p_value < 0.05:
    print('Отвергаем нулевую гипотезу.')
else:
    print('\n Не отвергаем нулевую гипотезу.')

# сравниваем не медиану, а среднее, потому что в значениях много нулей
print('\nОтносительная разница в среднем числе заказов группы B к группе А: ',
      "{0:.3f}".format(sample_b.mean() / sample_a.mean() - 1))
```

**Вывод** 

Видим, что p-value групп по "сырым" данным больше уровня статистической значимости (0.01679 против 0.05). Отвергаем нулевую гипотезу о том, что нет статистически значимых различий между группами.

Среднее число заказов на посетителя у группы В выше чем группы А - примерно на 14%.

### Статистическая значимость различий в среднем чеке заказа между группами по «сырым» данным

Посчитаем статистическую значимость различий в среднем чеке между группами без удаления аномальных данных.

**Нулевая гипотеза**: различий в среднем чеке между группами нет. 

**Альтернативная гипотеза**: различия в среднем чеке между группами есть. 

```python
sample_a = orders[orders['group']=='A']['revenue']
sample_b = orders[orders['group']=='B']['revenue']

p_value = stats.mannwhitneyu(sample_a, sample_b).pvalue

print('\np-value среднего чека по "сырым данным":', '{0:.5f}'.format(p_value))

if p_value < 0.05:
    print('\nОтвергаем нулевую гипотезу. ')
else:
    print('\nНе отвергаем нулевую гипотезу.')

# сравниваем не медиану, а среднее, потому что в значениях много нулей
print('\nОтносительная разница среднего чека группы B к группе А', 
      '{0:.3f}'.format(sample_b.mean()/sample_a.mean()-1)) 
```

**Вывод** 

Из полученных результатов видим, что p-value среднего чека групп по "сырым" данным значительно больше уровня статистической значимости (0.729 против 0.05). 
Мы не отвергаем нулевую гипотезу и не обнаруживаем статистически значимых различий между распределениями средних чеков у групп.

При этом средний чек группы В значительно выше среднего чека группы А - примерно на 26%. 

### Статистическая значимость различий в среднем количестве заказов на посетителя между группами по «очищенным» данным

Теперь посмотрим статистическую значимость в среднем числа заказов по данным, из которых исключены аномальные пользователи. 

Подготовим выборки количества заказов по пользователям по группам теста без учета аномальных пользователей. Затем применим статистический тест Манна-Уитни.

```python
# подготовим "очищенную" выборку числа заказов по пользователям по группам теста
# для тех пользователей, кто не совершил ни одной покупки поставим 0

# фильтр аномальных пользвоателей
orders_by_users_a_clean = orders_by_users_a[~orders_by_users_a['userId'].isin(abmormal_users)]

# число пользователей                                            
total_visitors_a = data_raw['visitors_per_day_a'].sum()

# число пользователей, которые что-то купили
buyers_a = len(orders_by_users_a_clean)

# заполним нулями для пользователей, которые ничего не купили                                            
zeros_a = pd.Series([0] * (total_visitors_a - buyers_a), name='orders')

                                            
# фильтр аномальных пользвоателей
orders_by_users_b_clean = orders_by_users_b[~orders_by_users_b['userId'].isin(abmormal_users)]                                           
total_visitors_b = data_raw['visitors_per_day_a'].sum()
buyers_b = len(orders_by_users_b_clean)                                            
zeros_b = pd.Series([0] * (total_visitors_b - buyers_b), name='orders')                                            
                                            
                                            
                                            
# объединим данные
sample_filter_a = pd.concat([orders_by_users_a_clean['orders'], zeros_a], ignore_index=True)   
sample_filter_b = pd.concat([orders_by_users_b_clean['orders'], zeros_b], ignore_index=True)                                             


# посчитаем p-value и относительное различие среднего числа заказов по очищенным данным 
                                            
p_value = stats.mannwhitneyu(sample_filter_a, sample_filter_b).pvalue                                             
                                            
print('\np-value по "очищенным данным":', 
      '{0:.5f}'.format(p_value))
                                            
                                            
if p_value < 0.05:
    print('Отвергаем нулевую гипотезу. ')
else:
    print('\nНе отвергаем нулевую гипотезу.')

                                            
# сравниваем не медиану, а среднее, потому что в значениях много нулей                                            
print('\nОтносительная разница среднего числа заказов группы В к группе А:',
      '{0:.3f}'.format(sample_filter_b.mean()/sample_filter_a.mean()-1))
```

**Вывод** 
По очищенным данным видим, что p-value значительно уменьшился в сравнении с сырыми данными (по сырым - 0.0168, по очищенным - 0.00626), и остался меньше уровня статистической значимости в 0.05. 

По очищенным данным среднее число заказов на посетителя группы В значительно выше, чем у группы А - на 18%, что даже больше, чем по "сырым" данным на 3%.

### Статистическая значимость различий в среднем чеке заказа между группами по «очищенным» данным

Теперь посмотрим, изменились ли результаты по очищенным данным по среднему чеку.

```python
# подготовим очищенные данные о стоимости заказов без аномальных пользователей
avg_order_filter_a = orders[(orders['group'] == 'A') & (~orders['visitorId'].isin(abmormal_users))]['revenue']

# skupina B bez abnormálních uživatelů
avg_order_filter_b = orders[(orders['group'] == 'B') & (~orders['visitorId'].isin(abmormal_users))]['revenue']


# посчитаем p-value и относительное различие


p_value = stats.mannwhitneyu(avg_order_filter_a, avg_order_filter_b).pvalue

print('p-value среднего чека по "очищенным" данным:', 
      '{0:.5f}'.format(p_value))

if p_value < 0.05:
    print('Отвергаем нулевую гипотезу. ')

else:
    print('\nНе отвергаем нулевую гипотезу.')
    
print('\nОтносительная разница разница среднего чека группы B к группе А:', 
      '{0:.3f}'.format(avg_order_filter_b.mean()/avg_order_filter_a.mean()-1)) 
```

**Вывод** 

Как видим, p-value значительно превышает уровень статистической значимости (0.79772 против 0.05). Мы не отвергаем нулевую гипотезу и не обнаруживаем статистически значимых различий между распределениями средних чеков у групп.

Значение p-value по "очищенным" данным практически не изменилось в сравнении с "сырыми" (по сырым - 0.72929, по очищенным - 0.79772).

При этом прирост в среднем чеке группы В вместо 25% изменился на отрицательный показатель в -2%, т.е. средние чеки по очищенным данным у обеих групп были практически равными.
Такие результаты соотносятся с графиком относительного различия среднего чека и подтверждают, что аномалии могут влиять на прочтение результатов анализа A/B тестирования.
